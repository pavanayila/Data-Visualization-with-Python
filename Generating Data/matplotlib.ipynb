```python
#Plotting a Simple Line Graph

import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]

plt.plot(squares)

plt.show()
```

```python
#Changing the Label Type and Graph Thickness

import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]

#the line width parameter controls the thick of the line that plot() generates

plt.plot(squares, linewidth = 5)

#Set chart title and label axes

plt.title("Square Numbers", fontsize = 24)

plt.xlabel("Value", fontsize = 14)

plt.ylabel("Square of Value", fontsize = 14)

plt.show()
```

```python
#Plotting a Series of Points with scatter()

x_values = [1, 2, 3, 4, 5]

y_values = [1, 4, 9, 16, 25]

plt.scatter(x_values, y_values, s=100)

#Set chart title and label axes.

plt.title("Square Numbers", fontsize = 24)

plt.xlabel("Value", fontsize = 14)

plt.ylabel("Square of Value", fontsize = 14)

#Set size of tick labels.

plt.tick_params(axis = 'both', which = 'major', labelsize = 14)

plt.show()
```

```python
#Calculating Data Automatically

import matplotlib.pyplot as plt

x_values = list(range(1, 1001))

y_values = [x**2 for x in x_values]

plt.scatter(x_values, y_values, s = 40)

#Set chart title and label axes.

plt.title("Square Numbers", fontsize = 24)

plt.xlabel("Value", fontsize = 14)

plt.ylabel("Square of Value", fontsize = 14)

#Set size of tick labels.

plt.tick_params(axis = 'both', which = 'major', labelsize = 14)

#Set the range for each axis.

plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
#Removing Outlines from Data Points

import matplotlib.pyplot as plt

x_values = list(range(1, 1001))

y_values = [x**2 for x in x_values]

#To remove the outlines around points, pass the arguement edgecolor = 'none'

#because the default blue dots with a black outline works well for plots with a few points but when plotting many points, the black outlines can blend together

plt.scatter(x_values, y_values, edgecolor = 'none', s = 40)

#Set chart title and label axes.

plt.title("Square Numbers", fontsize = 24)

plt.xlabel("Value", fontsize = 14)

plt.ylabel("Square of Value", fontsize = 14)

#Set size of tick labels.

plt.tick_params(axis = 'both', which = 'major', labelsize = 14)

#Set the range for each axis.

plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
#Defining Custom Colors

import matplotlib.pyplot as plt

x_values = list(range(1, 1001))

y_values = [x**2 for x in x_values]

#To change the color of the points, pass c to scatter() with name of a color to use

plt.scatter(x_values, y_values, c = 'red', edgecolor = 'none', s = 40)

plt.title("Square Numbers", fontsize = 24)

plt.xlabel("Value", fontsize = 14)

plt.ylabel("Square of Value", fontsize = 14)

plt.tick_params(axis = 'both', which = 'major', labelsize = 14)

plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
#Using a Colormap

import matplotlib.pyplot as plt

x_values = list(range(1, 1001))

y_values = [x**2 for x in x_values]

#we pass the list of y-values to c and then tell pyplot which colormap to use through the cmap arguement

plt.scatter(x_values, y_values, c = y_values, cmap = plt.cm.Blues, edgecolor = 'none', s = 40)

#The above line, colors the points with lower y-values light blue and the points with larger y-values darkblue

plt.title("Square Numbers", fontsize = 24)

plt.xlabel("Value", fontsize = 14)

plt.ylabel("Square of Value", fontsize = 14)

plt.tick_params(axis = 'both', which = 'major', labelsize = 14)

plt.axis([0, 1100, 0, 1100000])

plt.show()
```
